# Morphological Transformations
- Operations based on the shape of the image.
- Requires as input parameters:
  - The original image;
  - The kernel or structuring element;

## Erosion
- Makes the pixels 1 if all the pixels on the image are 1 on the surface of the kernel.

```python
import cv2
import numpy

from matplotlib import pyplot

original_image = cv2.imread("../../../data/images/i.png", 0)
kernel = numpy.ones((5, 5), numpy.uint8)
eroded_image = cv2.erode(original_image, kernel, iterations=1)

pyplot.imshow(original_image)
pyplot.title("Original image")
pyplot.show()

pyplot.imshow(eroded_image)
pyplot.title("Eroded image")
pyplot.show()
```

## Dilation
- Makes the pixels 1 if atleast one pixel on the image is 1 on the surface of the kernel.

```python
dilated_image = cv2.dilate(original_image, kernel, iterations=1)

pyplot.imshow(dilated_image)
pyplot.title("Dilated image")
pyplot.show()
```

## Opening
- Another name for erosion followed by dilation.

```python
opened_image = cv2.morphologyEx(original_image, cv2.MORPH_OPEN, kernel)

pyplot.imshow(opened_image)
pyplot.title("Opened image")
pyplot.show()
```

## Closing
- Another name for erosion after dilation.
- The opposite of opening, obviously.

```python
closed_image = cv2.morphologyEx(original_image, cv2.MORPH_CLOSE, kernel)

pyplot.imshow(closed_image)
pyplot.title("Closed image")
pyplot.show()
```

## Gradient
- The difference of erosion and dilation.
- The result looks like the outline of the object.

```python
gradient_image = cv2.morphologyEx(original_image, cv2.MORPH_GRADIENT, kernel)

pyplot.imshow(gradient_image)
pyplot.title("Gradient image")
pyplot.show()
```

## Top Hat
- The difference of original and opened image.

```python
top_hat_image = cv2.morphologyEx(original_image, cv2.MORPH_TOPHAT, kernel)

pyplot.imshow(top_hat_image)
pyplot.title("Top Hat image")
pyplot.show()
```

## Black Hat
- The difference of original and closed image.

```python
black_hat_image = cv2.morphologyEx(original_image, cv2.MORPH_BLACKHAT, kernel)

pyplot.imshow(black_hat_image)
pyplot.title("Black Hat image")
pyplot.show()
```

## Structuring Element
- Create manually the kernel.

```python
cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5, 5))
cv2.getStructuringElement(cv2.MORPH_CROSS, (5, 5))
```
